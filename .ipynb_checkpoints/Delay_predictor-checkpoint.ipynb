```python
import pandas as pd
datasetx = pd.read_csv('combined_filtered.csv',error_bad_lines=False)
datasetx.info()
```

```python
datasetx = datasetx.drop(['Dest_space','Origin_space','Unnamed: 0','OP_UNIQUE_CARRIER','ORIGIN','DEST'],axis=1)
```

```python
datasetx['TOTAL_DELAY'].fillna(-1000,inplace=True)
```

```python
delay_array = datasetx['TOTAL_DELAY']
```

```python
import numpy as np
class_label = [None] * 1727654

for i in range(len(delay_array)):
    if delay_array[i] ==0 :
        class_label[i]='Nd'
    elif delay_array[i]>0 and delay_array[i]<=15:
        class_label[i]='A'
    elif delay_array[i]>15 and delay_array[i]<=30:
        class_label[i]='B'
    elif delay_array[i]>30 and delay_array[i]<=45:
        class_label[i]='C'
    elif delay_array[i]>45 and delay_array[i]<=60:
        class_label[i]='D'
    elif delay_array[i]>60:
        class_label[i]='E'
    elif delay_array[i]==-1000:
        class_label[i]='CN'
        
```

```python
df=pd.DataFrame({'col':class_label})
```

```python
datasetx = pd.concat([datasetx, df],axis=1)
```
