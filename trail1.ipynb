```python
import pandas as pd

dataset = pd.read_csv('combined.csv',low_memory=False)
dataset.head()
```

```python
dataset = dataset.drop(['Unnamed: 0',
                        'ORIGIN_AIRPORT_ID',  
                        'DEST_AIRPORT_ID',
                        'DEST_CITY_MARKET_ID',
                        'Origin_id','Origin_station_name',
                        'Unnamed: 27',
                        'Dest_id',
                        'station_name'
                       ],axis=1)
```

```python
dataset['TOTAL_DELAY'] = dataset['DEP_DELAY_NEW']+dataset['ARR_DELAY_NEW']
dataset = dataset.drop(['DEP_DELAY_NEW',
                        'ARR_DELAY_NEW'
                       ],axis=1)
```

```python
dataset=dataset.drop(['Origin_weather_type','Dest_weather_type'],axis=1)
```

```python
dataset.to_csv('combined_filtered.csv')
```

```python
import pandas as pd

datasetx = pd.read_csv('fukced.csv', low_memory=False)
datasetx.info()
```

```python
datasetx.fillna(0,inplace=True)
```

```python
from sklearn.model_selection import StratifiedShuffleSplit
split=StratifiedShuffleSplit(n_splits=1, test_size=0.2,random_state=42)
for train_index, test_index in split.split(datasetx, datasetx["CANCELLED"]):
    strat_train_set=datasetx.loc[train_index]
    strat_test_set= datasetx.loc[test_index]
```

```python
xtrain= strat_train_set.drop("CANCELLED",axis=1)
ytrain= strat_train_set["CANCELLED"].copy()
```

```python
from sklearn import tree
model = tree.DecisionTreeClassifier()
model.fit(xtrain, ytrain)
```
